```python
import torch
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = torch.FloatTensor([2,3,5,7,11])
_, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(13, 5))
ax1.plot(x, list(range(len(x))), 'r*:')
ax2.plot(x, list(range(len(x))), 'gp-.')
ax3.plot(x, list(range(len(x))), 'bo--')
```
